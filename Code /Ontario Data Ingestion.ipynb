```python
import sqlalchemy as sq
import requests, json
import pandas as pd
```

```python
#assign the resource id, limit and api_url 
resource_id = '544a3f0a-aed3-4dec-ac29-d4acb3be9ccc'
limit= 2000
api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}'.format(resource_id,limit)
print(api_url)
```

```python
#use the get request to pull the data from the api
Treasury_data = requests.get(api_url)
Treasury_data
```

```python
#convert data to json
Treasury_data = Treasury_data.json()
```

```python
#view converted data
Treasury_data
```

```python
#get the help information in the data through dictionary key
Treasury_data['help']
```

```python
#get the records in the data
Treasury_data['result']['records']
```

```python
#convert data to dataframe
Treasury_infomation = pd.DataFrame(Treasury_data['result']['records'])
Treasury_infomation
```

```python
#insert code do data cleaning and engineering
Treasury_infomation.info()
```

```python
#Convert Salaray Paid and Taxable Benefits into float

Treasury_infomation['Salary Paid'] = Treasury_infomation['Salary Paid'].str.replace('[^\d.]', '', regex=True)
Treasury_infomation['Taxable Benefits'] = Treasury_infomation['Taxable Benefits'].str.replace('[^\d.]', '', regex=True)

# Step 2: Convert the columns to numeric data types
Treasury_infomation['Salary Paid'] = pd.to_numeric(Treasury_infomation['Salary Paid'], errors='coerce')
Treasury_infomation['Taxable Benefits'] = pd.to_numeric(Treasury_infomation['Taxable Benefits'], errors='coerce')

# Display the updated DataFrame
print(Treasury_infomation.info())
```

```python
Treasury_infomation
```

```python
#Find duplicate values
Treasury_infomation.duplicated().any()
```

```python
# Define the bin edges for salary and taxable benefits categories
salary_bins = [0, 50000, 100000, 150000, float('inf')]
benefits_bins = [0, 5000, 10000, 15000, float('inf')]

# Define the labels for salary and taxable benefits categories
salary_labels = ['Low', 'Medium', 'High', 'Very High']
benefits_labels = ['Low', 'Medium', 'High', 'Very High']

# Create new columns with the categories
Treasury_infomation['Salary Category'] = pd.cut(Treasury_infomation['Salary Paid'], bins=salary_bins, labels=salary_labels)
Treasury_infomation['Benefits Category'] = pd.cut(Treasury_infomation['Taxable Benefits'], bins=benefits_bins, labels=benefits_labels)

# Display the updated DataFrame
print(Treasury_infomation)
```
