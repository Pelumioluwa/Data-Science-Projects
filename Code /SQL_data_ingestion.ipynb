```python
#define a function to remove non digit character and character that aren't period
def remove(df,col):
    df[col] = df[col].str.replace('[^\d.]','',regex=True)
```

```python
#define function to get columns and dataframe
columns = ['Total Visits', 'Sameday Visits', 'Overnight Visits',
       'Total Expenditures', 'Sameday Expenditures', 'Overnight Expenditures',
       'Change in Total Visits', 'Change in Sameday Visits',
       'Change in Overnight Visits', 'Change in Total Expenditures',
       'Change in Sameday Expenditures', 'Change in Overnight Expenditures']

for i in columns:
    remove(outbound,i)

outbound
```

```python
#convert columns from object to numeric
outbound['Total Visits'] = pd.to_numeric(outbound['Total Visits'], errors='coerce')
outbound['Sameday Visits'] = pd.to_numeric(outbound['Sameday Visits'], errors='coerce')
outbound['Overnight Visits'] = pd.to_numeric(outbound['Overnight Visits'], errors='coerce')

outbound.info()
```

```python
#convert to float 
def tofloat(df,col):
    df[col] = df[col].astype(float)

columnname =['Total Expenditures', 'Sameday Expenditures', 'Overnight Expenditures',
       'Change in Total Visits', 'Change in Sameday Visits',
       'Change in Overnight Visits', 'Change in Total Expenditures',
       'Change in Sameday Expenditures', 'Change in Overnight Expenditures']
for i in columnname :
    tofloat(outbound,i)

outbound.info()
```

```python
#push cleaned and gotten data in postgres database (Pelumi)
#pandas dataframe to sql
inbound.to_sql(
name = 'inbound',
con = pelumi_db_engine,
schema= 'schulich',
if_exists ='replace',
index = False,
dtype = {'id': sa.types.INTEGER,
          'Year' :sa.types.INTEGER, 
          'Status' :sa.types.VARCHAR, 
          'Origin': sa.types.VARCHAR, 
          'Destination':sa.types.VARCHAR , 
          'Direction':sa.types.VARCHAR ,
          'Total Visits':sa.types.INTEGER, 
          'Sameday Visits':sa.types.INTEGER, 
          'Overnight Visits':sa.types.INTEGER,
          'Total Expenditures':sa.types.FLOAT, 
          'Sameday Expenditures':sa.types.FLOAT, 
          'Overnight Expenditures':sa.types.FLOAT,
          'Change in Total Visits':sa.types.FLOAT, 
          'Change in Sameday Visits':sa.types.FLOAT,
          'Change in Overnight Visits':sa.types.FLOAT, 
          'Change in Total Expenditures':sa.types.FLOAT,
          'Change in Sameday Expenditures':sa.types.FLOAT, 
         'Change in Overnight Expenditures':sa.types.FLOAT},
method = 'multi'
)
outbound.to_sql(
name = 'outbound',
con = pelumi_db_engine,
schema= 'schulich',
if_exists ='replace',
index = False,
dtype = {'id': sa.types.INTEGER,
          'Year' :sa.types.INTEGER, 
          'Status' :sa.types.VARCHAR, 
          'Origin': sa.types.VARCHAR, 
          'Destination':sa.types.VARCHAR , 
          'Direction':sa.types.VARCHAR ,
          'Total Visits':sa.types.INTEGER, 
          'Sameday Visits':sa.types.INTEGER, 
          'Overnight Visits':sa.types.INTEGER,
          'Total Expenditures':sa.types.FLOAT, 
          'Sameday Expenditures':sa.types.FLOAT, 
          'Overnight Expenditures':sa.types.FLOAT,
          'Change in Total Visits':sa.types.FLOAT, 
          'Change in Sameday Visits':sa.types.FLOAT,
          'Change in Overnight Visits':sa.types.FLOAT, 
          'Change in Total Expenditures':sa.types.FLOAT,
          'Change in Sameday Expenditures':sa.types.FLOAT, 
         'Change in Overnight Expenditures':sa.types.FLOAT},
method = 'multi'
)

tpi.to_sql(
name = 'tpi',
con = pelumi_db_engine,
schema= 'schulich',
if_exists ='replace',
index = False,
dtype = {'id': sa.types.INTEGER,
          'Year' :sa.types.INTEGER, 
          'Status' :sa.types.VARCHAR, 
          '%ChangeTPI Ontario':sa.types.FLOAT,
          'Price of Gasoline ($/L)':sa.types.FLOAT, 
          '%ChangeTPI US':sa.types.FLOAT,
          'Price of Gasoline (US$/Gal)':sa.types.FLOAT, 
          'Ontario TPI / CPI Ratio':sa.types.FLOAT,
          '%ChangeOntario TPI / CPI Ratio':sa.types.FLOAT, 
          'Ontario-US Relative TPI':sa.types.FLOAT,
          '%Change Ontario-US Relative TPI':sa.types.FLOAT},
method = 'multi'
)
```

```python
#connect to database with engine
with pelumi_db_engine.connect() as connection:
    inbound = pd.read_sql(
        sql='SELECT * FROM schulich.inbound',
        con = connection
    )

inbound
```

```python
#query the database
qwhere = '''
SELECT ib."Origin",
       ib."Destination",
       ib."Direction"
FROM schulich.inbound AS ib
WHERE ib."Origin" = 'Other Provinces'
'''
with pelumi_db_engine.connect() as connection:
    return_query = connection.execute(qwhere)
    samp = pd.DataFrame(return_query.fetchall(),columns=return_query.keys())

samp
```

```python
#using union. 
qwhere = ''' 
SELECT ib."Direction",
       ib._id,
       ib."Origin"
FROM schulich.inbound AS ib
WHERE ib."Year" = 2013

UNION

SELECT ob."Destination",
       ob._id,
       ob."Origin"
FROM schulich.outbound AS ob;
'''

with pelumi_db_engine.connect() as connection:
    return_query = connection.execute(qwhere)
    samp = pd.DataFrame(return_query.fetchall(),columns=return_query.keys())

samp
```
