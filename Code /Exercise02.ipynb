```python
#import libraries 
import sqlalchemy as sa
import requests,json
import pandas as pd
```

```python
#assign api key, resource id, limit and api_url for covid-19 vaccine
resource_id = '775ca815-5028-4e9b-9dd4-6975ff1be021'
covid_vac_api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}'.format(resource_id)
print(covid_vac_api_url)
```

```python
#use the api url to pull the data for covid-19 vaccine
covid_vac= requests.get(covid_vac_api_url)
covid_vac
```

```python
#convert inbound data to json
covid_vac= covid_vac.json()
covid_vac
```

```python
#get the records in the data
covid_vac['result']['records']
```

```python
#convert covid vaccine data to dataframe
covid_vac = pd.DataFrame(covid_vac['result']['records'])
```

```python
covid_vac
```

```python
covid_vac.info()
```

```python
#convert  columns to numeric 
covid_vac['At least one dose_cumulative'] = pd.to_numeric(covid_vac['At least one dose_cumulative'], errors='coerce')
covid_vac['Second_dose_cumulative'] = pd.to_numeric(covid_vac['Second_dose_cumulative'], errors='coerce')
covid_vac['fully_vaccinated_cumulative'] = pd.to_numeric(covid_vac['fully_vaccinated_cumulative'], errors='coerce')
covid_vac['third_dose_cumulative'] = pd.to_numeric(covid_vac['third_dose_cumulative'], errors='coerce')
covid_vac['Total population'] = pd.to_numeric(covid_vac['Total population'], errors='coerce')
```

```python
#convert to float
covid_vac['Percent_at_least_one_dose'] = covid_vac['Percent_at_least_one_dose'].astype(float)
covid_vac['Percent_fully_vaccinated'] = covid_vac['Percent_fully_vaccinated'].astype(float)
covid_vac['Percent_3doses'] = covid_vac['Percent_3doses'].astype(float)
```

```python
#convert to datetime
covid_vac['Date'] = pd.to_datetime(covid_vac['Date'])
```

```python
covid_vac.columns
```

```python
covid_vac = covid_vac.drop(columns=['_id', 'Second_dose_cumulative', 
       'third_dose_cumulative', 'Total population',
       'Percent_at_least_one_dose', 'Percent_fully_vaccinated',
       'Percent_3doses'], axis=1)
```
